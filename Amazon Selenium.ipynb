```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
```

```python
driverPath =  r"C:\\Users\\ELCOT\driver\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverPath)


driver.get("https://www.amazon.in/s?k=kurtis+set+for+women+latest+design&crid=3TP5QUP47GDSP&ref=nb_sb_ss_ts-doa-p_1_6")
name = driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']")
orgprice = driver.find_elements_by_xpath("//span[@class='a-price a-text-price']")
offerprice = driver.find_elements_by_xpath("//span[@class='a-price']")
lst = []
l1 = []
l2 = []

writer=pd.ExcelWriter(r"C:\\Users\\ELCOT"+"\\"+'amazon.xlsx',engine='xlsxwriter')

for i in name:
    lst.append(i.text)

for j in orgprice:
    l1.append(j.text)

for k in offerprice:
    l2.append(k.text)


df = pd.DataFrame(list(zip(lst,l1,l2)), columns = ['Product Name','Original Price','Offer Price'])
df.to_excel(writer,index = False)
writer.save()
writer.close()
print(df)
driver.close()
```
